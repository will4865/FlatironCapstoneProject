## Remove HTML tags

##### Made list with Beautiful Soup to iterate through of HTML stopwords to remove from scripts in Data Frame. 

```python
html_tag_list = []
html_tags_cleaned = []
html_page = requests.get('https://www.w3schools.com/TAGS/default.asp')
soup = BeautifulSoup(html_page.content, 'html.parser')
html_tags = list(soup.findAll('td'))
html_tags = html_tags[::2]
for i in html_tags:
    html_tag_list.append(str(i).split('.asp">')[1][:-9])
for i in html_tag_list:
    temp_list = []
    temp_list.append(i.replace('&lt;','<'))
    for i in temp_list:
        html_tags_cleaned.append(i.replace('&gt;','>'))
```

```python
html_tags = []
for i in html_tags_cleaned:
    temp_list = []
    temp_list.append(i[1:])
    for x in temp_list:
        html_tags.append('</' + x)
```

```python
html_tags_clean = html_tags_cleaned + html_tags
html_tags_clean
```

```python
html_tags_cleaned
open_tags = []
for i in html_tags_cleaned:
    open_tags.append(i[:-1])
```

```python
html_tags_clean = html_tags_clean + open_tags
```

```python
len(html_tags_clean)
```

```python
def clean_html_tags(script):
    for x in script:
        if x in html_tags_clean:
            script.replace(x, '')
```
