```python
letter_list = list(string.ascii_uppercase)
```

```python
letter_list.append('sign')
```

```python
subs_title_list = []
```

```python
for letter in letter_list:
    html_page = requests.get('https://subslikescript.com/movies_letter-{}'.format(letter))
    soup = BeautifulSoup(html_page.content, 'html.parser')
    script = list(soup.findAll(class_='scripts-list'))
    list_of_titles = str(script[0]).split('<a')[1:]
    for title in list_of_titles:
        subs_title_list.append(title.split('<li>')[1].replace('</li></a>\n','')[:-7])
```

```python
def subs_like_script_pull(dict_of_titles):
    for original_title, cleaned_title in dict_of_titles.items():
        driver = webdriver.Chrome('/Users/will4856/Downloads/chromedriver')
        driver.get("https://subslikescript.com/")
        elem = driver.find_element_by_xpath("/html/body/div/div/main/nav[1]/form/input")
        elem.clear()
        elem.send_keys(cleaned_title)
        elem.send_keys(Keys.RETURN)
        time.sleep(1)
        driver.find_element_by_xpath('/html/body/div/div/main/article/ul/a').click()
        BS_url = driver.current_url
        script = BS(BS_url, 'div', class_= 'full-script')
        index_at = films_at_large[films_at_large['title'] == original_title].index.tolist() 
        add_script_to_df(index_at, script, films_at_large)
        driver.close()
```

Pulled titles on first page of every letter in the alphabet

```python
scripts_com_list = []
for letter in letter_list:
    html_page = requests.get('https://www.scripts.com/scripts/{}'.format(letter))
    soup = BeautifulSoup(html_page.content, 'html.parser')
    titles = list(soup.findAll('td', class_='tal qx'))
    for title in titles:
        scripts_com_list.append(str(title).split('<strong>')[1].split('</strong>')[0])
```
